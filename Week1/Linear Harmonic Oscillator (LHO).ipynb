# Linear Harmonic Oscillator (LHO)

LHO equation:
    $ \frac{d^2x}{dt^2} + \omega^2x=0 $

#### LHO Posistion vs. Time graph

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def LHO (y,t,omega0):
    y1,y2=y
    return [y2,(-omega0**2)*y1]

# initial conditions
omega0=[0.5,1,2,4]
inits=[5,0] # initial position x0=5 and initial velocity v=0
t=np.linspace(0,20,1000) 

for index, value in enumerate(omega0):
    sol=integrate.odeint(LHO,inits,t,args=(value,))
    fig=plt.figure(1,figsize=(8,6), dpi=150)
    plt.subplot(2,2,index+1)
    plt.plot(t, sol[:,0], label='omega0='+str(value))
    plt.subplots_adjust(hspace=0.4, wspace=0.2)
    plt.xlabel('time [s]')
    plt.ylabel('x(t) [m]')
    plt.legend()
```

#### LHO Position vs. Velocity graph

```python
for index, value in enumerate(omega0):
    sol=integrate.odeint(LHO,inits,t,args=(value,))
    plt.figure(1,figsize=(8,6), dpi=150)
    plt.subplot(2,2,index+1)
    plt.plot(sol[:,0], sol[:,1], label='omega0='+str(value))
    plt.subplots_adjust(hspace=0.4, wspace=0.2)
    plt.xlabel('time [s]')
    plt.ylabel('velocity [m/s]')
    plt.legend()
```

#### LHO - Kinetic and Potential Energy vs. Time Graph

```python
init = [5, 0]
m=1 #mass of the oscillating body
k=np.power(omega0[0],2)*m #konst krutosti
sol=integrate.odeint(LHO,init,t,args=(omega0[0],))
Ek=0.5*m*np.power(sol[:,1],2)
Ep=0.5*k*np.power(sol[:,0],2)

fig = plt.figure(3,figsize=(8,6),dpi=150)
plt.plot(t,Ek,color='blue',linewidth=1.5,linestyle='-')
plt.plot(t,Ep,color='red',linewidth=1.5,linestyle='--')
plt.xlabel("time[s]")
plt.ylabel("Ek and Ep [J]")
plt.legend(["Ek","Ep"])
fig.savefig('Kinetic and potential energy vs time graph.png')
```

```python
fig = plt.figure(3,figsize=(8,6),dpi=150)
plt.plot(sol[:,0],Ek,color='blue',linewidth=1.5,linestyle='-')
plt.plot(sol[:,0],Ep,color='red',linewidth=1.5,linestyle='--')
plt.xlabel("position [m]")
plt.ylabel("Ek and Ep [J]")
plt.legend(["Ek","Ep"])
fig.savefig('Kinetic and potential energy vs position graph.png')
```
